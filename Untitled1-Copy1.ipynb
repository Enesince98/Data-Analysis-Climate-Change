```python
# -*- coding: utf-8 -*-
"""
Created on Tue Dec 14 22:35:41 2021

@author: i_hav
"""

import pandas as pd
import numpy as np
dF = pd.read_csv("ocean.csv") #reads file
desc = dF.describe() #creates the table
dF["Date Time"]=pd.to_datetime(dF["Date Time"])
dF=dF.set_index(dF["Date Time"])
dF.drop(["Date Time"],axis=1,inplace=True)
dF.drop(["Tpot (K)"],axis=1,inplace=True)
```

```python
dF.head()
```

```python
dF.describe()
```

```python
dFrollingH=dF.resample("H").mean()
rollingHdesc=dFrollingH.describe()
dFstdH=dF.resample("H").std()
dFrollingH.to_csv("ocean_Hourly_Mean.csv")
dFstdH.to_csv("ocean_Hourly_std.csv")
```

```python
dFrollingH.describe()
```

```python
dFrollingD=dF.resample("D").mean()
rollingDdesc=dFrollingD.describe()
dFstdD=dF.resample("D").std()
dFrollingD.to_csv("ocean_Daily_Mean.csv")
dFstdD.to_csv("ocean_Daily_std.csv")
```
